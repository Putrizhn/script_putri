MERGE ALL Faktor Iklim Data + tb

```python
df_tb_iklim = df_tb.merge(df_rh, on=['KABKOT', 'month', 'year'], how='left').merge(df_tp, how='left')
df_tb_iklim
```

```python
df_tb_iklim = df_tb_iklim.sort_values(by=[ 'KABKOT', 'year', 'month'])
df_tb_iklim = df_tb_iklim[['KABKOT', 'month', 'year', 'time', 'longitude', 'latitude', 'tb', 'rh', 'tp']]
df_tb_iklim
```
