```python
import pandas as pd
import xarray as xr
import numpy as np
import xlrd
```

open all csv

```python
df_ku = pd.read_csv('/Users/Jia/Downloads/data_skripsi/uji_data/merged_data_KU.csv')
df_iklim = pd.read_csv(r'/Users/Jia/Downloads/data_skripsi/uji_data/merged_data_iklim.csv')
```

merapihkan data

```python
df_ku = df_ku.drop(['Unnamed: 0',], axis=1)
```

```python
df_ku
```

MERGE ALL KU Iklim Data + tb

```python
df_alldata = df_ku.merge(df_iklim, on=['KABKOT', 'month', 'year','tb'], how='left')
```

```python
df_alldata
```

```python
df_alldata = df_alldata.drop(['time_y', 'longitude_y', 'latitude_y'], axis=1)
df_alldata.rename(columns= {'time_x':'time', 'longitude_x':'longitude', 'latitude_x':'latitude'}, inplace=True)
df_alldata
```

```python
df_alldata = df_alldata.sort_values(by=[ 'KABKOT', 'year', 'month'])
df_alldata = df_alldata[['KABKOT', 'month', 'year', 'time', 'longitude', 'latitude', 'tb', 'pm2p5', 'pm10', 'so2', 'rh', 'tp']]
df_alldata
```
