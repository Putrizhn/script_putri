```python
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr_ku = pd.read_csv("/Users/Jia/Downloads/data_skripsi/uji_data/merged_data_KU.csv")
```

```python
corr_ku['time']= pd.to_datetime(corr_ku['time'])
corr_ku = corr_ku.drop(['time'], axis=1)
corr_ku.info()
```

```python
#filter corr_ku based on year
ku_2020 = corr_ku[corr_ku['year'] == 2020]
ku_2021 = corr_ku[corr_ku['year'] == 2021]
ku_2022 = corr_ku[corr_ku['year'] == 2022]
```

```python
# extract column names starting from the 7th column (index 6) onward
i_2020 = (ku_2020.columns[6:])
i_2021 = (ku_2021.columns[6:])
i_2022 = (ku_2022.columns[6:])
print(i_2020,i_2021,i_2022)
```

CORR KU 2020 JAKARTA

```python
plt.figure(figsize=(10,5))
plt.title("Variables Correlation in Jakarta 2020")
cmap = sns.diverging_palette(260,-10,s=50, l=50, n=6, as_cmap=True) #pewarnaan
sns.heatmap(ku_2020[i_2020].corr(), square=True, annot=True, linewidths=.5, cmap=cmap, vmin=-1, vmax=1)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Variables Correlation in Jakarta 2021")
cmap = sns.diverging_palette(260,-10,s=50, l=50, n=6, as_cmap=True) #pewarnaan
sns.heatmap(ku_2021[i_2021].corr(), square=True, annot=True, linewidths=.5, cmap=cmap, vmin=-1, vmax=1)
plt.show()
```
